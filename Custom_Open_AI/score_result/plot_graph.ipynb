```python
from matplotlib import pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline, splrep
import numpy as np

def plot_result(file_nm, graph):
    with open (file_nm, "r") as txt:
        result = txt.read().split(",")
        result = list(filter(None, result))
        result = list(map(int, result))
    txt.close()
    plt.ylim(min(result)- 3, max(result) + 3)
    plt.plot(result)
    plt.savefig(graph)
    plt.show()

    print(f"the maximum score is {max(result)}")
    print(f"the minimum score is {min(result)}")
    print(f"the average score is {sum(result)/len(result)}")
```

```python
# The result for DQN with out proritized replay
plot_result("score_DQN_ERM.txt", "DQN_ERM_score.png")
```

```python
# The result for PPO
plot_result("score_PPO.txt", "PPO_score.png")
```
